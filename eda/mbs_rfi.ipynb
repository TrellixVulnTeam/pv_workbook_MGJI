```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import datetime
import matplotlib.patches as mpatches
```

```python
y1 = [40,32,29,20,10,9,5,5,3,3]
y2 = [30,30,28,24,21,20,20,18,10,9]
y3 = [11,5,4,6,2,3,2,1,3,1]
y4 = [62,50,49,45,38,27,22,19,18,15]
x = np.arange(len(y1))
```

```python
plt.stackplot(x,y1,y2,y3,y4, colors=['mediumaquamarine','lightsalmon','indianred','plum'])
plt.title('MBS Solar RFIs')
plt.xlabel('2019')
plt.ylabel('Open RFIs')
plt.xticks(x,['Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct'])
purple_patch = mpatches.Patch(color='mediumaquamarine',label='Milestones')
red_patch = mpatches.Patch(color='lightsalmon',label='API')
blue_patch = mpatches.Patch(color='indianred',label='8760')
plum_patch = mpatches.Patch(color='plum',label='Metadata')
plt.legend(handles=[purple_patch,red_patch,blue_patch,plum_patch], loc='upper left')
plt.figure(figsize=(5,8))
plt.show()
```
