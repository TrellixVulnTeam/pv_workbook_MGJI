```python
def clean_plant(str):
    str_list = str.split()
    del str_list[:2]
    clean_plt = ' '.join(str_list)
    return clean_plt
df['Clean Plant'] = df['Plant'].apply(clean_plant)
df['Plant'] = df['Clean Plant']
```
