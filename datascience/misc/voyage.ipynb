```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('data.xls')
```

```python
df.columns
```

```python
countries = df[['Spain / Uruguay', 'Portugal / Brazil', 'Great Britain','Netherlands', 'U.S.A.', 'France', 'Denmark / Baltic']].transpose()
```

```python
sns.heatmap(countries,  cmap="YlGnBu")
```

```python
brazil = df['Portugal / Brazil']
us = df['U.S.A.']
brits = df['Great Britain']
france = df['France']
```

```python
# sns.lineplot(x=range(0,len(brazil)),y=brazil)
plt.plot(brazil, label='Portugal')
plt.plot(us, label='United States')
plt.plot(brits,label='Great Britain')
plt.plot(france, label='France')
plt.title('Trans-Atlantic Slave Trade')
plt.xlabel('Time Period : 1500-1875')
plt.ylabel('Voyages (Thousands)')
plt.legend()
```
